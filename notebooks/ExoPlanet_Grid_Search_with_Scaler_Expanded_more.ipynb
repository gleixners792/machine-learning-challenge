```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
#df.head()
df
```

```python
#df.info()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_fpflag_co',  'koi_fpflag_nt',  'koi_fpflag_ss',  'koi_model_snr',  'koi_prad',  'koi_duration_err2',  'koi_prad_err2',  'koi_duration_err1',  'koi_steff_err1',  'koi_fpflag_ec',  'koi_prad_err1',  'koi_steff_err2',  'koi_time0bk_err2',  'koi_period',  'koi_time0bk_err1',  'koi_duration',  'koi_period_err1',  'koi_depth',  'koi_period_err2',  'koi_impact',  'koi_insol_err1',  'koi_insol',  'koi_teq',  'koi_insol_err2',  'koi_time0bk',  'koi_depth_err1',  'koi_depth_err2',  'ra',  'dec',  'koi_impact_err1',  'koi_impact_err2',  'koi_slogg_err2',  'koi_kepmag',  'koi_srad_err1']]
selected_features
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
disposition = df["koi_disposition"]
disposition_names = ["CANDIDATE", "CONFIRMED", "FALSE POSITIVE"]
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(selected_features, disposition, random_state=42)
```

```python
X_train.head()
```

```python
y_train
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
```

# Pre-processing
Scale the data using the MinMaxScaler and perform some feature selection

```python
from sklearn.preprocessing import MinMaxScaler
X_minmax = MinMaxScaler().fit(X_train)

X_train_scale = X_minmax.transform(X_train)
X_test_scale = X_minmax.transform(X_test)
```

```python
X_train_scale
```

# Hyperparameter Tuning
Use GridSearchCV to tune the model's parameters

```python
# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [449, 450, 451],
              'gamma': [0.0000001, 0.000001, 0.00001]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Fit the model using the grid search estimator. 
# This will take the SVC model and try each combination of parameters
grid.fit(X_train_scale, y_train)
```

```python
print(f"Training Data Score: {grid.score(X_train_scale, y_train)}")
print(f"Testing Data Score: {grid.score(X_test_scale, y_test)}")
```

```python
# List the best parameters for this dataset
print(grid.best_params_)
```

```python
# List the best score
print(grid.best_score_)
```

```python
# Best index for the model to impact the model.
print(grid.best_index_)
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test_scale)
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=disposition_names))
```
